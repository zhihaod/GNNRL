```python

def spread_on_Gs(G_list,ini_num,times,N,seeds=None,tau=1.2,gamma=1):
    com_nodes = (G_list[-1].nodes)
    report_times = np.linspace(0,600,81)

    initial_infecteds = sample(list(com_nodes),ini_num)
   # initial_infecteds = seeds
    #res = []
    for i in range(len(G_list)):
        S,I,R = simulate_spread_times(G_list[i],tau, gamma,times,initial_infecteds,report_times)
        res = ([report_times,S/N,I/N,R/N])
        
    return np.array(res)
                     

def simulate_spread_times(G,tau, gamma,times,ini_infecteds,report_times):
    Ssum = np.zeros(len(report_times))
    Isum = np.zeros(len(report_times))
    Rsum = np.zeros(len(report_times))
    iterations = times
    for counter in range(iterations):
        t, S, I, R = EoN.fast_SIR(G, tau, gamma, initial_infecteds = ini_infecteds)
        newS, newI, newR = EoN.subsample(report_times, t, S, I, R)    
        Ssum += newS
        Isum += newI
        Rsum += newR
    Save = Ssum / float(iterations)
    Iave = Isum / float(iterations)
    Rave = Rsum / float(iterations)
    return Save,Iave,Rave
        
```

```python
def get_reduced_graphs(graph_dict,selected_nodes):
    r_graphs = []
    for g_index in range(1):
        graphs = []
        g = graph_dic[g_index].g.copy()
        for i in selected_nodes[0][g_index]:
            g.remove_node(i)
        g_rein = g.copy()
        graphs.append(g_rein)

        g = graph_dic[g_index].g.copy()
        deg_dict = dict(g.degree())
        keys = sorted(deg_dict, key=lambda key: deg_dict[key])               
        g.remove_nodes_from(keys[-len(results[0][g_index]):])
        g_degree = g.copy()
        graphs.append(g_degree)
        
        g = graph_dic[g_index].g.copy()
        for _ in range(len(selected_nodes[0][g_index])):
            i = choice(list(g.nodes))
            g.remove_node(i)    
        g_random = g.copy()
        graphs.append(g_random)
        
        r_graphs.append(graphs)
    return r_graphs


def cal_edges(graphs):

    e_num = [0,0,0]
    for index in range(len(graphs)):
        e_num[0] += nx.number_of_edges(graphs[index][0])
        e_num[1] += nx.number_of_edges(graphs[index][1])
        e_num[2] += nx.number_of_edges(graphs[index][2])
    e_num[0] = e_num[0] / len(graphs)
    e_num[1] = e_num[1] / len(graphs)
    e_num[2] = e_num[2] / len(graphs)
    return e_num

def cal_shortest_path(graphs):

    e_num = [0,0,0]
    for index in range(len(graphs)):
        e_num[0] += cal_shortest_path_(graphs[index][0])
        e_num[1] += cal_shortest_path_(graphs[index][1])
        e_num[2] += cal_shortest_path_(graphs[index][2])
    e_num[0] = e_num[0] / len(graphs)
    e_num[1] = e_num[1] / len(graphs)
    e_num[2] = e_num[2] / len(graphs)
    return e_num


def cal_shortest_path_(G_o,p=1):
    largest_cc = max(nx.connected_components(G_o), key=len)
    G = G_o.subgraph(largest_cc).copy() 
    
    Nodes = G.nodes()
    num = math.ceil(p * nx.number_of_nodes(G))
    set_node = sample(Nodes,num)
    paths = []
    for t in set_node:
        path_node = nx.shortest_path_length(G,target=t)
        paths.append(path_node)
    pd_paths = pd.DataFrame(paths)
    shortes_p = pd_paths.mean(axis=1).mean()
    return shortes_p



def cal_l_component(graphs):

    e_num = [0,0,0]
    for index in range(len(graphs)):
        e_num[0] += len(max(nx.connected_components(graphs[index][0]), key=len))
        e_num[1] += len(max(nx.connected_components(graphs[index][1]), key=len))
        e_num[2] += len(max(nx.connected_components(graphs[index][2]), key=len))
    e_num[0] = e_num[0] / len(graphs)
    e_num[1] = e_num[1] / len(graphs)
    e_num[2] = e_num[2] / len(graphs)
    return e_num



def cal_epid_threshold(graphs):
    e_num = [0,0,0]
    for index in range(len(graphs)):
        e_num[0] += cal_epid_threshold_(graphs[index][0])
        e_num[1] += cal_epid_threshold_(graphs[index][1])
        e_num[2] += cal_epid_threshold_(graphs[index][2])
    e_num[0] = e_num[0] / len(graphs)
    e_num[1] = e_num[1] / len(graphs)
    e_num[2] = e_num[2] / len(graphs)
    return e_num    

def cal_epid_threshold_(G):
    ret = {}
    largest_cc = max(nx.connected_components(G), key=len)
    G_cc = G.subgraph(largest_cc).copy()
    degrees = [k for (n, k) in G_cc.degree]
    ret["k"]  = np.array(degrees).mean()
    ret["k2"] = np.array(degrees).var() + np.array(degrees).mean()**2
    return 1 / (ret["k2"]/ret["k"] - 1)

def cal_num_of_components(graphs):
    e_num = [0,0,0]
    for index in range(len(graphs)):
        e_num[0] += nx.number_connected_components(graphs[index][0])
        e_num[1] += nx.number_connected_components(graphs[index][1])
        e_num[2] += nx.number_connected_components(graphs[index][2])
    e_num[0] = e_num[0] / len(graphs)
    e_num[1] = e_num[1] / len(graphs)
    e_num[2] = e_num[2] / len(graphs)
    return e_num        



def cal_connectivity(graphs):

    e_num = [0,0,0]
    for index in range(len(graphs)):
        e_num[0] += nx.average_node_connectivity(graphs[index][0])
        e_num[1] += nx.average_node_connectivity(graphs[index][1])
        e_num[2] += nx.average_node_connectivity(graphs[index][2])
    e_num[0] = e_num[0] / len(graphs)
    e_num[1] = e_num[1] / len(graphs)
    e_num[2] = e_num[2] / len(graphs)
    return e_num
```
