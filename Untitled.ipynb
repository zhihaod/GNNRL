```python
import EoN
import networkx as nx

def runSIR(G,times,ini_infecteds,report_times=np.linspace(0,10,201),tau=1.2, gamma=1 ):
    #report_times = np.linspace(0,15,101)
    Ssum = np.zeros(len(report_times))
    #Isum = np.zeros(len(report_times))
    #Rsum = np.zeros(len(report_times))
    iterations = times
    for counter in range(iterations):
        t, S, I, R = EoN.fast_SIR(G, tau, gamma, initial_infecteds = ini_infecteds)
        newS, newI, newR = EoN.subsample(report_times, t, S, I, R)    
        Ssum += newS
       # Isum += newI
       # Rsum += newR
    Save = Ssum / float(iterations)
    #Iave = Isum / float(iterations)
    #Rave = Rsum / float(iterations)
    return Save[-1]
```

```python
def runSIR_1(G,times,ini_infecteds,report_times=np.linspace(0,10,201),tau=1.2, gamma=1 ):
    #report_times = np.linspace(0,15,101)
    #Ssum = np.zeros(len(report_times))
    #Isum = np.zeros(len(report_times))
    #Rsum = np.zeros(len(report_times))
    iterations = times
    Isum = 0
    for counter in range(iterations):
        newI = EoN.get_infected_nodes(G, tau, gamma, initial_infecteds=ini_infecteds)
           
        #Ssum += newS
        Isum += len(newI)
       # Rsum += newR
    #Save = Ssum / float(iterations)
    Iave = Isum / float(iterations)
    #Rave = Rsum / float(iterations)
    return G.number_of_nodes() - Iave
```

```python
import random
from copy import deepcopy
def runIC (G, S, p=0.1 ):
    ''' Runs independent cascade model.
    Input: G -- networkx graph object
    S -- initial list of vertices
    p -- propagation probability
    Output: T -- resulted influenced set of vertices (including S)
    '''
    
    T = deepcopy(S) 
    for u in T: 
         for v in G[u]: 
            w = 1
            if v not in T and random.random() < 1 - (1-p)**w:
                T.append(v)
    return T

def runIC_repeat(G, S, p=0.01, sample=1000):
    infl_list = []
    for i in range(sample):
        T = runIC(G, S, p=p)
        influence = len(T)
        infl_list.append(influence)
    infl_mean = np.mean(infl_list)
    infl_std = np.std(infl_list)

    return infl_mean, infl_std 
```
